## Third: Evaluate Data Quality Issues in the Data Provided

```python
import json
import pandas as pd

# Load JSON data from file
file_path = r'D:\Assignment\users.json'

# Initialize an empty list to store JSON objects
data_list = []

# Read the file line by line and concatenate JSON objects
with open(file_path) as f:
    for line in f:
        data_list.append(json.loads(line))

# Convert JSON list to DataFrame
df = pd.DataFrame(data_list)

# Write DataFrame to CSV
csv_file_path = r'D:\Assignment\users.csv'
df.to_csv(csv_file_path, index=False)

# Read CSV into pandas DataFrame
users_df = pd.read_csv(csv_file_path)

# Display DataFrame
users_df.head()
```

```python

# Load JSON data from file
file_path = r'D:\Assignment\receipts.json'

# Initialize an empty list to store JSON objects
data_list = []

# Read the file line by line and concatenate JSON objects
with open(file_path) as f:
    for line in f:
        data_list.append(json.loads(line))

# Convert JSON list to DataFrame
df = pd.DataFrame(data_list)

# Write DataFrame to CSV
csv_file_path = r'D:\Assignment\receipts.csv'
df.to_csv(csv_file_path, index=False)

# Read CSV into pandas DataFrame
receipts_df = pd.read_csv(csv_file_path)

# Display DataFrame
receipts_df.head()
```

```python

# Load JSON data from file
file_path = r'D:\Assignment\brands.json'

# Initialize an empty list to store JSON objects
data_list = []

with open(file_path) as f:
    for line in f:
        data_list.append(json.loads(line))

# Convert JSON list to DataFrame
df = pd.DataFrame(data_list)

# Write DataFrame to CSV
csv_file_path = r'D:\Assignment\brands.csv'
df.to_csv(csv_file_path, index=False)

# Read CSV into pandas DataFrame
brands_df = pd.read_csv(csv_file_path)

# Display DataFrame
brands_df.head()
```
